```python
!date
df_null_ = df_.groupby('plm_community_id').agg(
    community_size = ('accession', len),
    mean_pLDDT = ('af2_mean_pLDDT', np.mean),
    null_Pfam = ('dbReference_Pfam_Id', lambda a: a.isnull().all()),
    null_InterPro = ('dbReference_InterPro_Id', lambda a: a.isnull().all()),
    null_PROSITE = ('dbReference_PROSITE_Id', lambda a: a.isnull().all()),
    null_SUPFAM = ('dbReference_SUPFAM_Id', lambda a: a.isnull().all()),
)
!date
df_null_.head()
```

```python
df_null_.query('mean_pLDDT > 70 & 5 < community_size & community_size < 100 & null_Pfam & null_InterPro & null_PROSITE & null_SUPFAM')\
.sort_values('mean_pLDDT', ascending=False)
```
